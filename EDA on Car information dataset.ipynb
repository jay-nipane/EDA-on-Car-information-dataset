#### Handling the outliers using flooring and capping method

```python
# flooring and capping method

outlier_list = []

# sorting data
data['horsepower'] = sorted(data['horsepower'])

# Finding first and third qurtile
q1 = np.percentile(data['horsepower'], 25)
q3 = np.percentile(data['horsepower'], 75)
print("The Val of Q1 and Q2",q1, q3)

IQR = q3-q1

lwr_bound = q1-(1.5*IQR)
upr_bound = q3+(1.5*IQR)

    # print("The lower & Upper Bound",lwr_bound, upr_bound)
    
for i in data['horsepower']: 
    if (i<lwr_bound or i>upr_bound):
        data['horsepower'] = data['horsepower'].replace(i,np.median(data['horsepower']))
```

```python
# verifying outlyer

# Initialize a flag to check if there are any outliers
outlier_present = False

for i in data['horsepower']: 
    if (i<lwr_bound or i>upr_bound):
        print(i)
        outlier_present = True

# If the flag remains False, no outliers were found
if not outlier_present:
    print("There are no outliers.")
```

#### Handling the outliers usingTrimming and Capping method

```python
# Handling the outliers
acc = data['acceleration']

tenth_percentile = np.percentile(acc, 10)
ninetieth_percentile = np.percentile(acc, 90)
    
print('10th,tenth_percentile',tenth_percentile)
print('90th,tenth_percentile',ninetieth_percentile)
    
b = np.where(acc<tenth_percentile, tenth_percentile, acc)

b1 = np.where(b>ninetieth_percentile, ninetieth_percentile, b)

data['acceleration'] = b1 
```
