#### Handling the outliers using flooring and capping method

```python
# flooring and capping method

outlier_list = []

# sorting data
data['horsepower'] = sorted(data['horsepower'])

# Finding first and third qurtile
q1 = np.percentile(data['horsepower'], 25)
q3 = np.percentile(data['horsepower'], 75)
print("The Val of Q1 and Q2",q1, q3)

IQR = q3-q1

lwr_bound = q1-(1.5*IQR)
upr_bound = q3+(1.5*IQR)

    # print("The lower & Upper Bound",lwr_bound, upr_bound)
    
for i in data['horsepower']: 
    if (i<lwr_bound or i>upr_bound):
        data['horsepower'] = data['horsepower'].replace(i,np.median(data['horsepower']))
```

<div class = 'alter alter-block alert-info' >
    <h4>Observation</h4>
    <p><i>- The provided code is implementing the flooring and capping method to handle outliers in the 'horsepower' column of the data.</i></p>
    <p><i>- The flooring and capping method is a simple approach to handle outliers by replacing them with the median value</i></p>
</div>

```python
# verifying outlyer

# Initialize a flag to check if there are any outliers
outlier_present = False

for i in data['horsepower']: 
    if (i<lwr_bound or i>upr_bound):
        print(i)
        outlier_present = True

# If the flag remains False, no outliers were found
if not outlier_present:
    print("There are no outliers.")
```

<div class = 'alter alter-block alert-info' >
    <h4>Observation</h4>
    <p><i>- The provided code is verifying the presence of outliers in the 'horsepower' column of the data using the lower bound (lwr_bound) and upper bound (upr_bound) values calculated previously with the flooring and capping method.</i></p>
    <p><i>- If it finds any outliers, it prints those outlier values. If no outliers are found, it informs that there are no outliers in the 'horsepower' column.</i></p>
</div>

#### Handling the outliers usingTrimming and Capping method

```python
# Handling the outliers
acc = data['acceleration']

tenth_percentile = np.percentile(acc, 10)
ninetieth_percentile = np.percentile(acc, 90)
    
print('10th,tenth_percentile',tenth_percentile)
print('90th,tenth_percentile',ninetieth_percentile)
    
b = np.where(acc<tenth_percentile, tenth_percentile, acc)

b1 = np.where(b>ninetieth_percentile, ninetieth_percentile, b)

data['acceleration'] = b1 
```

<div class = 'alter alter-block alert-info' >
    <h4>Observation</h4>
    <p><i>- The provided code is handling the outliers in the 'acceleration' column of the data using a capping method</i></p>

### <em>Radar Chart (Spider Chart):</em>

```python
# Data for the radar chart
categories = ['USA', 'JAPAN', 'EUROPE']
values = [245, 79, 68]

# Duplicate the first value to complete the circle (closed shape)
values += values[:1]

# Calculate the angles for each category to plot on the polar axes
angles = [n / float(len(categories)) * 2 * np.pi for n in range(len(categories))]
angles += angles[:1]

# Create the polar plot
plt.polar(angles, values)

# Fill the area inside the plot with color to create a filled radar chart
plt.fill(angles, values, alpha=0.5)

# Set the category names as labels on the x-axis
plt.xticks(angles[:-1], categories)

# Add a title to the radar chart
plt.title('Radar Chart')

# Display the radar chart
plt.show()
```

<div class = 'alter alter-block alert-info' >
    <h4>Observation</h4>
<p><i>The resulting radar chart will display the values for each category, forming a closed shape with lines connecting the points for each category. </i></p>
